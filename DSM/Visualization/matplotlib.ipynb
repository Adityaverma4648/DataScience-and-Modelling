```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./Toyota.csv",index_col=0,na_values=['??','????'])
df["MetColor"].replace("0.",0 ,inplace=True)
df["MetColor"].replace(1.,1 ,inplace=True )
df["MetColor"].value_counts().index[0]
df["MetColor"].fillna(df["MetColor"].value_counts().index[0],inplace=True)
df["FuelType"].value_counts().index[0]
df["FuelType"].fillna(df["FuelType"].value_counts().index[0],inplace=True)
df["KM"].fillna(df["KM"].median(),inplace=True)
df["HP"].fillna(df["HP"].mean(),inplace=True)
df["Age"].fillna(df["Age"].mean(),inplace=True)
print(df.isna().sum())
print(df.info())
```

Visualization using matplotlib

001 .    ScatterPlot

```python
plt.scatter(df["Age"],df["Price"],c ="red")
plt.title("Relation btw Age and Price")
plt.xlabel("Age(months)")
plt.xlabel("Price(Euros)")
plt.show()
```

002.    Histogram

```python
plt.hist(df["KM"],color ="black",edgecolor = "white",bins = 5)
plt.title("Freq dist of df[`KM`]")
plt.xlabel("KM")
plt.xlabel("Freq")
plt.show()
```

003.  Bar plot

```python
Count = df["FuelType"].value_counts()
myCount = [1277 ,144,15]
print(myCount)
myLabel = ["petrol", "diesel","CNG"]
index = np.arange(len(myLabel))
print(index)


plt.bar(index , myCount , color=["red","blue","green"])
plt.title("bar plot")
plt.xlabel("Fuel Types")
plt.ylabel("Frequency")
plt.xticks(index,myLabel,rotation = 90)
```
