Doing some Visualixation using SeaBorn
SeaBorn consists of
 1> Box And Whiskers plot
 2> pairwise Plot
 3> Scatter plot
 4> Bar Plot
 5> Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd

df = pd.read_csv("./Toyota.csv",index_col=0,na_values=["??","????"])
df.dropna(axis = 0,inplace=True)

df.info()
```

PLotting Scatter plot in seaBorn - plotting relation between Price and Age 

Scatter plot in SeaBorn it has a default param set to (True) i.e fit_reg = True , it estimates and plots a regression model  relating both the axis

```python
sns.set(style="darkgrid")
sns.regplot(x = df["Age"], y = df["Price"])
```

with fit_reg = False

```python

sns.set(style="darkgrid")
sns.regplot(x = df["Age"], y = df["Price"] ,fit_reg=False)
```

Some variations in Scatter plot

```python
sns.set(style="darkgrid")
sns.regplot(x = df["Age"], y = df["Price"] ,fit_reg=False, marker="*")
```

Adding hue and showing distribution in unique values of FuelType 

```python
sns.lmplot(x= "Age" , y = "Price" ,data = df , fit_reg=False , hue='FuelType' ,legend=True)
```

Plotting Histogram plot for df["Age"]


```python
sns.displot(df["Age"])
```

similar to the scatter plot it also has regression model showing distribution of values {**kde**}

```python
sns.displot(df["Age"] ,kde = False)
```

Similar to matplotlib we can mention bins in histogram of seaborn

```python
sns.displot(df["Age"] ,kde = False,bins=5)
```

Results a relation btw count and the Age 

Bar PLot in Seaborn between FuelType and the catgory Automatic

```python
sns.countplot(x = "FuelType" ,data = df,hue="Automatic" )
```

Box wishkers plot 

```python
sns.boxplot(x = "FuelType" ,y = "Price" , data =df )
```

Potting boxWhiskers for categories of a column types

```python
sns.boxplot(x = "FuelType" , y = "Price",data = df , hue="Automatic")
```
